```
# -*- coding: utf-8 -*-
"""
Created on Mon Aug 10 10:46:08 2020

@author: Jacobo Paredes
"""

import re
path =  "/content/"
corpus = {}

class doctos:
    def __init__(self,no_docto,wordpos):
        self.doctos = {}
        self.doctos[1] = no_docto
        self.freq = [1]
        self.wordpos = [[wordpos]]
    def update(self,no_docto,wordpos,doccnt):
        if no_docto in self.doctos.values():
            self.freq[-1] += 1
            self.wordpos[-1].append(wordpos)
        else:
            self.doctos[doccnt+1] = no_docto
            self.freq.append(1)
            self.wordpos.append([1])

class token:
    def __init__(self,word,no_docto,wordpos):
        self.word = word
        self.doccnt = 1
        self.freqcnt = 1
        self.doctos = doctos(no_docto,wordpos)
    def update(self,no_docto,wordpos):
        self.freqcnt += 1
        self.doctos.update(no_docto,wordpos,self.doccnt)
        self.doccnt = len(self.doctos.doctos)

text_files = open(path+"listacorpus.txt",encoding = "ISO-8859-1")
text_names = text_files.read()
text_names = text_names.split('\n')
text_files.close()

for i in range(len(text_names)):
    docto_file = open(path+text_names[i]+".txt",encoding = "ISO-8859-1")
    docto_text = docto_file.read()
    docto_text = docto_text.lower()
    docto_text = re.sub("[^a-záéíóúüñ]"," ",docto_text)
    docto_text = re.sub("\s+"," ",docto_text)
    docto_words = docto_text.split()
    for j in range(len(docto_words)):
        if docto_words[j] in corpus:
            corpus[docto_words[j]].update(i+1,j+1)
        else:
            corpus[docto_words[j]] = token(docto_words[j],i+1,j+1)

print(corpus["hermosa"].doctos.freq)
print(corpus["hermosa"].doctos.doctos)
```
